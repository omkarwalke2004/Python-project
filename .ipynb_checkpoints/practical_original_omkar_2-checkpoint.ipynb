```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(df)
```

```python
numeric_cols = ['sub1', 'sub3', 'sub4', 'sub5']

# Plot box plots for all numeric columns
plt.figure(figsize=(15, 10))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col], color='skyblue')
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
Q1 = df['sub1'].quantile(0.25)
Q3 = df['sub1'].quantile(0.75)
IQR = Q3 - Q1

Lower_limit = Q1 - 1.5 * IQR
Upper_limit = Q3 + 1.5 * IQR

print(f'Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}, Lower_limit = {Lower_limit}, Upper_limit = {Upper_limit}')
```

```python
outliers = df[(df['sub1'] < Lower_limit) | (df['sub1'] > Upper_limit)]

print(f"\nOutliers in '{'sub1'}':")
print(outliers[['sub1']])
```

```python
outliers=[]
for i in df.sub1:
  if i<Lower_limit or i>Upper_limit:
       outliers.append(i)
print("outliers are",outliers)
```

```python
print("\nBefore Transformation Skewness:", df['percentage'].skew())
```

```python
sns.kdeplot(df.percentage)
```

```python
df['percentage_sqrt'] = np.sqrt(df['percentage'] - df['percentage'].min() + 1)
print("After sqrt transformation skewness:", df['percentage_sqrt'].skew())
```

```python
df['percentage_log'] = np.log(df['percentage'] - df['percentage'].min() + 1)
print("After log transformation skewness:", df['percentage_log'].skew())
```

```python
sns.kdeplot(df['percentage_log'])  # or 'percentage_sqrt', etc.
plt.title("KDE Plot After Transformation")
plt.show()
```

```python
sns.kdeplot(df['percentage_sqrt'])  # or 'percentage_sqrt', etc.
plt.title("KDE Plot After Transformation")
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')
df['yeojohnson_percentage'] = pt.fit_transform(df[['percentage']])
print("Yeo-Johnson Skewness:", pd.Series(df['yeojohnson_percentage']).skew())
```

```python
sns.kdeplot(df['yeojohnson_percentage'])  # or 'percentage_sqrt', etc.
plt.title("KDE Plot After Transformation")
plt.show()
```
